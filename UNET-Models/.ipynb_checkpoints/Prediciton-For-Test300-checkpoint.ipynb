```python
import os
import random
import numpy as np
import json
import h5py

np.random.seed(42)
random.seed(42)
import matplotlib.pyplot as plt
```

```python
from Prediction_Maker import *
```

```python
epochs = {
    "F606w": "049",
    "F625w": "049",
    "F775w": "056",
    "F814w": "013",
    "F850lp": "068",
    "MultiChannel":"050",
    "MultiEncoder":"092"
}
```

```python
split_data = np.load('../Preprocessing-Dataset-UNET/Train-Validation-Split-By-Region.npz')
testing = split_data['Test_dataset']
```

```python
testing = np.sort(testing)
```

```python
luminosity_test_input = {}

with h5py.File('../Preprocessing-Dataset-UNET/full-dataset-300.h5', 'r') as f:
    
    mass_test_target = f["Simulation_Mass_Maps"][testing].reshape((len(testing)*30,128,128))
    
    redshift_test_input = f["redshift"][testing].flatten()

    for filt in filter_names:
        luminosity_test_input[filt] = f[f"Simulation_Luminosity_Maps/{filt}"][testing].reshape((len(testing)*30,128,128))
```

```python
redshift_test_input = redshift_test_input.astype(np.float32)
```

```python
predict = {f_i:[] for f_i in Model_Names}
```

```python
for i in Model_Names[0:5]:
    ml_model = f"results_{i}/checkpoints_{i}/epoch{epochs[i]}.h5"
    
    predict[i] = Single_Prediction(ml_model, luminosity_test_input, redshift_test_input,filter_type=i)
```

```python
model = f"results_MultiChannel/checkpoints_MultiChannel/epoch{epochs['MultiChannel']}.h5"
predict['MultiChannel']=Multi_Channel_Prediction(model, luminosity_test_input, redshift_test_input)
```

```python
model = f"results_MultiEncoder/checkpoints_Multi/unet-multiview-film-continuous_epoch{epochs['MultiEncoder']}.h5"
predict['MultiEncoder']=MultiEncoder_Mass_Prediction(model, luminosity_test_input, redshift_test_input)
```

```python
with h5py.File("The300-Test-Results_LOG.h5", "w") as f:



    f.create_dataset("T300_Mass_Maps", data=mass_test_target)
    
    grp_unet = f.create_group("T300_Unet_Mass_Maps")
    
    for filt, maps in predict.items():
        data = np.stack(maps, axis=0)  
        grp_unet.create_dataset(filt, data=data)
    
    grp_lum = f.create_group("T300_Luminosity_Maps")
    for filt, maps in luminosity_test_input.items():
        data = np.stack(maps, axis=0)  
        grp_lum.create_dataset(filt, data=data)


```
